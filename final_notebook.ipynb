```python
import pandas as pd
import pyterrier as pt
```

```python
if not pt.started():
    pt.init()
```

```python
files = pt.io.find_files("./www.washington.edu/")
files
```

```python
files = ["./www.washington.edu"]
indexer = pt.FilesIndexer(
    "./uw_index",
    meta={"docno": 20, "filename": 1024, "title": 1024},
    meta_tags={"title": "title"},
    overwrite=True
)
indexref = indexer.index(files)
```

```python
index = pt.IndexFactory.of(indexref)
print (index.getCollectionStatistics().toString())
```

```python
tf_idf = pt.BatchRetrieve(index, wmodel="TF_IDF")
```

```python
queries = pd.DataFrame([["q1", "dawgs"], ["q2", "mathematics"]], columns=["qid", "query"])
queries.head()
```

```python
results = tf_idf.transform(queries)
```

```python
for i in range(len(results)):
    doc_id = results.docid[i]
    filename = index.getMetaIndex().getItem("filename", doc_id)  # Get the filename

    # Read the file's content
    try:
        with open(filename, 'r') as file:
            print(file.read())  # Output the content of the document
    except Exception as e:
        print(f"Error reading file {filename}: {e}")
```
